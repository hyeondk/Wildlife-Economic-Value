#### **1. 직교설계 카드 Data Preprocessing**

##### (1) 더미 변수 부여
* 변수 PRICE : Categorical Variable (Dummy)

```python
# Dictionary
ols_data_dict = {}

# 변수 1 : 실물소재 분류생태정보
ols_data_dict['BIO1'] = [] # 속수준 동정
ols_data_dict['BIO2'] = [] # 종수준 동정
ols_data_dict['BIO3'] = [] # 종수준과 서식환경 정보

# 변수 2 : 소재 희소성
ols_data_dict['SCA1'] = [] # 낮음
ols_data_dict['SCA2'] = [] # 중간
ols_data_dict['SCA3'] = [] # 높음

# 변수 3 : 전통지식 정보 제공
ols_data_dict['KNO1'] = [] # 미제공
ols_data_dict['KNO2'] = [] # 문헌정보
ols_data_dict['KNO3'] = [] # 문헌정보와 구전정보

# 변수 4 : 소재 유용성
ols_data_dict['USE1'] = [] # 미제공
ols_data_dict['USE2'] = [] # 공개 정보 제공
ols_data_dict['USE3'] = [] # 미공개 정보 제공

# 변수 5 : 추출물 제조 유형
ols_data_dict['MET1'] = [] # 비표준
ols_data_dict['MET2'] = [] # 표준

# 변수 6 : 추출물 분석 정보
ols_data_dict['ANA1'] = [] # 미제공
ols_data_dict['ANA2'] = [] # 제공

# 변수 7 : 가격(1 vial당 기준)
ols_data_dict['PRICE1'] = [] # 30,000원
ols_data_dict['PRICE2'] = [] # 300,000원
ols_data_dict['PRICE3'] = [] # 2,000,000원
```

```python
for i in range(len(card_data)) :
    
    # 변수 1 : 실물소재 분류생태정보
    if card_data['실물소재 분류생태정보'][i] == '속수준 동정':
        ols_data_dict['BIO1'].append(1)
        ols_data_dict['BIO2'].append(0)
        ols_data_dict['BIO3'].append(0)
    elif card_data['실물소재 분류생태정보'][i] == '종수준 동정':
        ols_data_dict['BIO1'].append(0)
        ols_data_dict['BIO2'].append(1)
        ols_data_dict['BIO3'].append(0)
    else:
        ols_data_dict['BIO1'].append(0)
        ols_data_dict['BIO2'].append(0)
        ols_data_dict['BIO3'].append(1)
    
    # 변수 2 : 소재 희소성
    if card_data['소재 희소성'][i] == '낮음':
        ols_data_dict['SCA1'].append(1)
        ols_data_dict['SCA2'].append(0)
        ols_data_dict['SCA3'].append(0)
    elif card_data['소재 희소성'][i] == '중간':
        ols_data_dict['SCA1'].append(0)
        ols_data_dict['SCA2'].append(1)
        ols_data_dict['SCA3'].append(0)
    else:
        ols_data_dict['SCA1'].append(0)
        ols_data_dict['SCA2'].append(0)
        ols_data_dict['SCA3'].append(1)
    
    # 변수 3 : 전통지식 정보 제공
    if card_data['전통지식 정보 제공'][i] == '미제공':
        ols_data_dict['KNO1'].append(1)
        ols_data_dict['KNO2'].append(0)
        ols_data_dict['KNO3'].append(0)
    elif card_data['전통지식 정보 제공'][i] == '문헌정보':
        ols_data_dict['KNO1'].append(0)
        ols_data_dict['KNO2'].append(1)
        ols_data_dict['KNO3'].append(0)
    else:
        ols_data_dict['KNO1'].append(0)
        ols_data_dict['KNO2'].append(0)
        ols_data_dict['KNO3'].append(1)
    
    # 변수 4 : 소재 유용성
    if card_data['소재 유용성'][i] == '미제공':
        ols_data_dict['USE1'].append(1)
        ols_data_dict['USE2'].append(0)
        ols_data_dict['USE3'].append(0)
    elif card_data['소재 유용성'][i] == '공개 정보 제공':
        ols_data_dict['USE1'].append(0)
        ols_data_dict['USE2'].append(1)
        ols_data_dict['USE3'].append(0)
    else:
        ols_data_dict['USE1'].append(0)
        ols_data_dict['USE2'].append(0)
        ols_data_dict['USE3'].append(1)
    
    # 변수 5 : 추출물 제조 유형
    if card_data['추출물 제조 유형'][i] == '비표준':
        ols_data_dict['MET1'].append(1)
        ols_data_dict['MET2'].append(0)
    else:
        ols_data_dict['MET1'].append(0)
        ols_data_dict['MET2'].append(1)
    
    # 변수 6 : 추출물 분석 정보
    if card_data['추출물 분석 정보'][i] == '미제공':
        ols_data_dict['ANA1'].append(1)
        ols_data_dict['ANA2'].append(0)
    else:
        ols_data_dict['ANA1'].append(0)
        ols_data_dict['ANA2'].append(1)
    
    # 변수 7 : 가격(1 vial당 기준)
    if card_data['가격'][i] == 30000:
        ols_data_dict['PRICE1'].append(1)
        ols_data_dict['PRICE2'].append(0)
        ols_data_dict['PRICE3'].append(0)
    elif card_data['가격'][i] == 300000:
        ols_data_dict['PRICE1'].append(0)
        ols_data_dict['PRICE2'].append(1)
        ols_data_dict['PRICE3'].append(0)
    else:
        ols_data_dict['PRICE1'].append(0)
        ols_data_dict['PRICE2'].append(0)
        ols_data_dict['PRICE3'].append(1)

# Dictionary -> Data Frame
ols_data_df = pd.DataFrame(ols_data_dict)
```

#### **2. 천연물 본설문 응답 Data Preprocessing**

##### (1) Conjoint Analysis(CA)를 통한 응답자 필터링
- Step 1 : 동일한 Rank 값 제외
- Step 2 : [검증용] 켄달타우 < 0 or 검증용 켄달타우 < 0.4 or '.' 제외
- Step 3 : [추정용(18개 CARD)] 피어슨 < 0.6147 or 켄달타우 < 0.4 제외
- Program : SPSS

##### (2) 각 카드에 대한 구매 가능성 응답 부여

```python
# List
ols_data = []

# 응답자 번호
card_index = file_data['REP'].to_list()

for i in range(len(card_index)):
    temp = ols_data_df.copy() # Data Frame 복사
    temp['Y'] = file_data.iloc[i, 1:].copy().to_numpy() # 구매 가능성(0~10)
    temp['INDEX'] = [card_index[i] for _ in range(18)] # 응답자 번호
    temp['CARD'] = [j+1 for j in range(18)] # 직교설계 카드 번호
    ols_data.append(temp.copy())

# Result
ols_data = pd.concat(ols_data)

# Data Frame Column 위치 변경
col1 = ols_data.columns[-2:].to_list()
col2 = ols_data.columns[:-2].to_list()
new_col_name = col1 + col2
ols_data = ols_data[new_col_name]
```

#### **3. 천연물 본설문 응답 OLS**

##### (1) 모든 더미 변수(dummy variable) 대입한 OLS

```python
# Independent variable : BIO1 ~ PRICE3(속성 및 수준)
ols_data.columns.to_list()[2:-1]
```

```python
# OLS
# Dependent variable : Y
# Independent variable : BIO1 ~ PRICE3(속성 및 수준)
nat_ols = ols('Y ~ ' + ' + '.join(ols_data.columns.to_list()[2:-1]), data=ols_data).fit()
nat_ols.summary()
```

##### (2) 각 더미 변수(dummy variable)에서 1개씩 제외한 모든 경우의 수의 OLS

```python
X_case = []
for BIO in ['BIO1', 'BIO2', 'BIO3']:
    for SCA in ['SCA1','SCA2', 'SCA3']:
        for KNO in ['KNO1','KNO2', 'KNO3']:
            for USE in ['USE1','USE2', 'USE3']:
                for MET in ['MET1', 'MET2']:
                    for ANA in ['ANA1', 'ANA2']:
                        for PRICE in ['PRICE1','PRICE2', 'PRICE3']:
                            X_case.append(f'Y ~ {BIO} + {SCA} + {KNO} + {USE} + {MET} + {ANA} + {PRICE}')

# 앞 부분만 확인
X_case[:5]
```

```python
# List 생성
# 1번째 내용 : 모든 더미 변수(dummy variable)를 대입한 OLS
nat_ols_all_case = [nat_ols.params.iloc[1:]]

for model in X_case:
    ols_fit = ols(model, data=ols_data).fit() # Fitting OLS
    nat_ols_all_case.append(ols_fit.params.iloc[1:].copy())

nat_ols_all_case = pd.concat(nat_ols_all_case, axis=1).T
nat_ols_all_case[nat_ols_all_case.isna()] = 0.0
```

#### **부록 : SPSS Utility 추정과 비교**

##### **Utility 산출 과정**
- **(1) 속성 변환**
     - (1-i) DISCRETE 자료 -> 각 속성별로 변수 배정
          - ex. POTENTIAL, DATA -> POTENTIAL1, POTENTIAL2, DATA1, DATA2, DATA3  
       
     <br>
     
     - (1-ii) LESS 자료 -> 각 속성별로 정수 값 배정
          - ex. PRICE의 40k, 100k, 300k -> 1, 2, 3  
<br>

- **(2) DISCRETE 자료 속성의 변수 중 한개씩 제외 후 OLS수행**
     - (2-i) 이후 OLS에서 제외된 변수들에게는 beta(회귀계수)값으로 0 배정
          - 변환된 데이터에서 [POTENTIAL2, TYPE2, INFODB2, DATA2, DATA3, SERVICE2, PRICE]를 독립변수로 OLS
          
          <br>  
          
          |CONSTANT|POTENTIAL1|POTENTIAL2|TYPE1|TYPE2|INFODB1|INFODB2|DATA1|DATA2|DATA3|SERVICE1|SERVICE2|PRICE|
          |:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
          |B0|0|B1|0|B2|0|B3|0|B4|B5|0|B6|B7|
  
<br>

- **(3) DISCRETE 자료들 중에서 속성별 beta의 합이 0이 되지 않는다면, 그 평균으로 보정값 산출**
     |POTENTIAL|TYPE|INFODB|DATA|SERVICE|
     |:---:|:---:|:---:|:---:|:---:|
     |[0+B1≠0] (0+B1)/2 = C1|[0+B2≠0] (0+B2)/2 = C2|[0+B3≠0] (0+B3)/2 = C3|[0+B4+B5≠0] (0+B4+B5)/3 = C4|[0+B6≠0] (0+B6)/2 = C5|
<br>

- **(4) 모든 속성의 보정값은 CONSTANT(상수)의 beta와 합산하고, 속성별로 각자의 보정값을 뺄셈**
     |CONSTANT|POTENTIAL1|POTENTIAL2|TYPE1|TYPE2|INFODB1|INFODB2|DATA1|DATA2|DATA3|SERVICE1|SERVICE2|PRICE|
     |:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
     |B0|0|B1|0|B2|0|B3|0|B4|B5|0|B6|B7|
     |+C1+C2+C3+C4+C5|-C1|-C1|-C2|-C2|-C3|-C3|-C4 |-C4|-C4|-C5|-C5||
